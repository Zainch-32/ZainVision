```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

k = sns.load_dataset("titanic")
k.groupby(["who","embark_town"]).survived.mean()
```

```python
# Renaming the dataframe rows
k.rename({'age':'AGE','pclass':'PCLASS','who':'WHO'},axis=1,inplace=True)
k.head()
```

```python
#k["WHO"] = k["WHO"]<400
sns.set_style("whitegrid")
sns.pairplot(data=k,hue="AGE",size=3)
#sns.swarmplot(data=k,x="sex",y="AGE",hue="embarked",color="red")
plt.xlabel("Nasha")
plt.show()
```
